```python
file_list = df_img.iloc[:,0].tolist()
print(len(file_list))
error_file = ['ciaa199.pdf-001-a.png' , 'ciaa199.pdf-001-b.png', 'ciaa199.pdf-001-c.png']
for er in error_file:
    file_list.remove(er)
print(len(file_list))
```

```python
clean_data_list = os.listdir(directory)
clean_data_list
```

```python
import imageio
import imgaug as ia
%matplotlib inline
data = os.path.join(directory,clean_data_list[1])
image = imageio.imread(data)

print("Original:")
ia.imshow(image)
```

```python
from imgaug import augmenters as iaa
ia.seed(4)

clahe = iaa.CLAHE()
image_aug = clahe(image=image)

print("Augmented:")
ia.imshow(image_aug)
```

```python
function_name = "CLAHE"
aug_dir = "augmented"

if not os.path.exists(aug_dir):
    os.makedirs(aug_dir)

save_dir = os.path.join(aug_dir, function_name)
    
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

data = clean_data_list[1]
data_name, ext = os.path.splitext(data)
data_path = os.path.join(directory, data)
image = imageio.imread(data_path)

if function_name == "CLAHE":
    function = iaa.CLAHE()
    
image_aug = function(image=image)
newPath = os.path.join(save_dir, data_name+f"_{function_name}"+ext)
imageio.imsave(newPath, image_aug)
```

```python
function_name = "CLAHE"
aug_dir = "augmented"
if not os.path.exists(aug_dir):
        os.makedirs(aug_dir)

save_dir = os.path.join(aug_dir, function_name)

if not os.path.exists(save_dir):
    os.makedirs(save_dir)

for data in clean_data_list:
    data_name, ext = os.path.splitext(data)
    data_path = os.path.join(directory, data)
    image = imageio.imread(data_path)

    if function_name == "CLAHE":
        function = iaa.CLAHE()

    image_aug = function(image=image)
    newPath = os.path.join(save_dir, data_name+f"_{function_name}"+ext)
    imageio.imsave(newPath, image_aug)
```
